```python
import pandas as pd
from pandasql import sqldf
df = pd.read_csv("https://covid.ourworldindata.org/data/owid-covid-data.csv")
```

```python
sqldf("select * from df limit 10")
```

```python
sqldf ("SELECT location,new_cases FROM df limit 10")
```

```python
sqldf("SELECT location,new_cases FROM df WHERE location='Uruguay' AND new_cases is NOT NULL")
```

```python
sqldf("SELECT location,new_cases,date FROM df WHERE location='Uruguay' limit 10 ")
```

```python
sqldf("SELECT location,new_cases,date FROM df WHERE location='Uruguay' AND new_cases IS NOT NULL ORDER BY date DESC limit 10 ")
```
